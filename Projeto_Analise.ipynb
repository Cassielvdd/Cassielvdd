```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px



df = pd.read_csv('2019.csv')

rank = df['Overall rank'][:20]
country = df['Country or region'][:20]
score = df['Score'][:20]

plt.barh(rank, score, color=['#FFA500', '#008000'])


for i in range(len(rank)):
    plt.text(score[i], rank[i], str(score[i]), ha='right', va='center')
    plt.text(score[i]  - 6.5, rank[i], str(df['Country or region'][i]), color='white', ha='left', va='center')


plt.gca().invert_yaxis()


plt.title('Felicidade no Mundo (Top 20 Países)')
plt.xlabel('Pontuação')
plt.ylabel('Rank Geral')


plt.show()
```

```python
import pandas as pd
import plotly.express as px


df = pd.read_csv("2019.csv")


fig = px.choropleth(
    df,
    locations="Country or region",
    locationmode="country names",
    color="Score",
    hover_name="Country or region",
    title="Felicidade no Mundo por País (2019)",
    color_continuous_scale="thermal",
)


fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("2019.csv")
media_gdp = df["GDP per capita"].mean()
media_social_support = df["Social support"].mean()
plt.title("GPD per Capita e Suporte Social")
labels = ["GDP per capita", "Suporte Social"]
valores = [media_gdp, media_social_support]
plt.pie(valores, labels=labels, autopct="%1.1f%%")
plt.axis("equal")
plt.show()
```
